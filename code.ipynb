```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

CSV = "myerson8.csv"
COL1 = "CLferror"
COL5 = "CRferror5"

df = pd.read_csv(CSV)

if COL1 not in df.columns or COL5 not in df.columns:
    missing = [c for c in (COL1, COL5) if c not in df.columns]
    print(f"Missing columns in {CSV}: {missing}", file=sys.stderr)
    raise SystemExit(1)

a = df[COL1].dropna()
b = df[COL5].dropna()

if a.empty and b.empty:
    print("Both columns are empty after dropping NaN.", file=sys.stderr)
    raise SystemExit(1)

# use common bin edges so histograms align
combined = pd.concat([a, b], ignore_index=True)
bins = np.histogram_bin_edges(combined, bins=50)

plt.figure(figsize=(8,5))
plt.hist(a, bins=bins, color='tab:blue', alpha=0.6, label=COL1)
plt.hist(b, bins=bins, color='tab:orange', alpha=0.6, label=COL5)
plt.legend()
plt.xlabel("Value")
plt.ylabel("Count")
plt.title(f"Histogram: {COL1} (blue) vs {COL5} (orange)")
plt.grid(alpha=0.25, linestyle='--')
plt.xlim(-1, 1)
plt.savefig("hist_CLferror_CRferror5.png", dpi=200)
plt.show()
```
